```python
nl = '\n'

print('By Deni Kusuma and Tania sari Dewi', nl)

a = int(input('Enter a number :')) #input pertama a berupa integer
b = int(input('Enter b number :')) #input pertama b berupa integer
c = int(input('Enter c number :')) #input pertama c berupa integer

val = [a,b,c]  #setiap input pada array val

def largest_value(a,b,c): #define setiap value
  if a > b and a > c: #jika a lebih dari b dan a lebih dari c return a
    return a
  elif b > a and b > c : #jika b lebih dari a dan b lebih dari c return b
    return b
  else: #jika c lebih besar dari kedua nilai tersebut return c
    return c

print('-------------------------------------------------')
print('In array list: ', val, nl)
print('Total Value in Array: ', len(val), nl)
print('The largest value is: ',largest_value(a,b,c))

```
